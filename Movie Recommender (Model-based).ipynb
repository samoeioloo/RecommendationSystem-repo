```python
#SVD Matrix Factorization
import pandas as pd
import numpy as np
import sklearn
from sklearn.decomposition import TruncatedSVD
```

```python
#prepare data
columns = ['user_id', 'item_id', 'rating', 'timestamp']
frame = pd.read_csv("C:/Users/P/OneDrive/Desktop/Dev Projects/Recommendation System ML AI/ml-100k/u.data", sep='\t', names=columns)
frame.head()
```

```python
#import item attributes
columns = ['item_id', 'movie title', 'release date', 'video release date', 'IMDb URL', 'unknown', 'Action',
          'Adventure', 'Animation', 'Children', 'Comedy', 'Crime', 'Documentary',
          'Drama', 'Fantasy', 'Film-Noir', 'Horror',
          'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
movies = pd.read_csv("C:/Users/P/OneDrive/Desktop/Dev Projects/Recommendation System ML AI/ml-100k/u.item", sep='|', names=columns, encoding='latin-1')
```

```python
movies.head()
```

```python
#combine the two dataframes into one
combinedMoviesData = pd.merge(frame, movies, on='item_id')
combinedMoviesData.head()
```

```python
#which movie has the most ratings?
#group combined df's by item id then countnumber of ratings given to each,then sort
combinedMoviesData.groupby('item_id')['rating'].count().sort_values(ascending=False).head()
```

```python
#find name of item_id 50 by filtering by boolean values
filter = combinedMoviesData['item_id']==50
combinedMoviesData[filter]['movie title'].unique()
```

```python
#Building a Utility Matrix - matrix contains a value for each movie and each user
ratingCrossTab = combinedMoviesData.pivot_table(values='rating', index='user_id', columns='movie_title', fill_value=0)
ratingCrossTab.head()
```
